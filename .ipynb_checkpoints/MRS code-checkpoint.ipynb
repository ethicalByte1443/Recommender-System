```python
import json
```

```python
def convert(list):
    new = []
    list = json.loads(list)
    for i in list:
        new.append(i['name'])
    return new
```

```python
def convert3(obj):
    list = json.loads(obj)
    new = []
    for i in list:
        new.append(i['name'])
    return new[0:3]
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies
```

```python
test = movies.iloc[0].crew
```

```python
def convert_crew_data(obj):
    list = json.loads(obj)
    new = []
    for i in list:
        if i['job'] == 'Director':
            new.append(i['name'])
            break
    return new
```

```python
print(convert_crew_data(test))
```

```python
movies['crew'] = movies['crew'].apply(convert_crew_data)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies
```

```python
movies['genres'] =  movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] =  movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] =  movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] =  movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast']+ movies["crew"]
```

```python
new_movies_df = movies[['movie_id', 'title', "tag"]]
```

```python
new_movies_df['tag'] = new_movies_df['tag'].apply(lambda x:" ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv
```

```python
vectors = cv.fit_transform(new_movies_df['tag']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('aseeming')
```

```python
new_movies_df['tag'] = new_movies_df['tag'].apply(stem)
```

```python
new_movies_df
```

```python
vectors = cv.fit_transform(new_movies_df['tag']).toarray()
```

```python
vectors
```

```python
for i in cv.get_feature_names_out():
    print(i)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity  = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_movies_df[new_movies_df['title'] == movie].index[0]     
    distances = similarity[movie_index]
    
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_movies_df.iloc[i[0]])
```

```python
recommend('Newlyweds')
```
